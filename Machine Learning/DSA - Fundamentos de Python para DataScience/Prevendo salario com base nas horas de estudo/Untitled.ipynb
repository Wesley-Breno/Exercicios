```python
import numpy as np
import pandas as pd  # Criar tabelas
import matplotlib.pyplot as plt  # Criar graficos
import seaborn as sns  # Criar graficos
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dataset.csv')
```

<h2>Preparacao dos dados</h2>

```python
x = np.array(df['horas_estudo_mes'])
```

```python
x = x.reshape(-1, 1)
```

```python
y = df['salario']
```

```python
plt.scatter(x, y, color = "blue", label = "Dados Reais Historicos")
plt.xlabel('Horas de estudo')
plt.ylabel('Salario')
plt.legend()
plt.show()
```

```python
# Dividir dados em treinamentos e testes
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
x_treino.shape
```

```python
x_teste.shape
```

```python
y_treino.shape
```

```python
y_teste.shape
```

<h2>Criando modelagem preditiva</h2>

```python
# Cria o modelo de regressao linear simples
model = LinearRegression()
```

```python
# Treina o modelo
model.fit(x_treino, y_treino)
```

```python
# Visualiza a reda de regressao linear (previsoes) e os dados mai usados no treinamento
plt.scatter(x, y, color = 'blue', label = 'Dados reais historicos')
plt.plot(x, model.predict(x), color = 'red', label = 'Reta de regressao com previsoes do modelo')
plt.xlabel('Horas de estudo')
plt.ylabel('Salario')
plt.legend()
plt.show()
```

```python
# Avalia o modelo nos dados de teste
score = model.score(x_teste, y_teste)
print(f'Coeficiente R^2: {score:.2f}')
```

```python
# Intercepto - parametro w0
model.intercept_
```

```python
# Slope - parametro w1
model.coef_
```

<h2>Usando Modelo</h2>

```python
# Exemplo 1
# Define um novo valor para horas de estudo
horas_estudo_novo = np.array([[48]])

# Fazendo previsao com o modelo treinado
salario_previsto = model.predict(horas_estudo_novo)

# Mostrando resultado
print(f'Se voce estudar cerca de {horas_estudo_novo[0][0]} horas por mês, seu salario pode ser igual a {salario_previsto}')
```
