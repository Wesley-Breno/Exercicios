<h1>Separating predictive data from classifiers</h1>

```python
x_houses = base_houses.iloc[:,5:6].values  # Taking footage of houses
x_houses
```

```python
y_houses = base_houses.iloc[:,2].values
y_houses
```

<h1>Separating training and testing data</h1>

```python
from sklearn.model_selection import train_test_split
```

```python
x_houses_train, x_houses_test, y_houses_train, y_houses_test = train_test_split(
    x_houses, 
    y_houses, 
    test_size=0.3, 
    random_state=0
)
```

```python
x_houses_train.shape, y_houses_train.shape
```

```python
x_houses_test.shape, y_houses_test.shape
```

<h1>Training the algorithm with separate data</h1>

```python
from sklearn.linear_model import LinearRegression
```

```python
simple_regressor_houses = LinearRegression()
simple_regressor_houses.fit(x_houses_train, y_houses_train)
```

<h1>Showing model performance</h1>

```python
#b0
simple_regressor_houses.intercept_
```

```python
#b1
simple_regressor_houses.coef_
```

```python
simple_regressor_houses.score(x_houses_train, y_houses_train)
```

```python
simple_regressor_houses.score(x_houses_test, y_houses_test)
```

```python
predicts = simple_regressor_houses.predict(x_houses_train)
predicts
```
