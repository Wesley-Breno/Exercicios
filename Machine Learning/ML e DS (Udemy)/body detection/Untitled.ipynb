<h1>Importing necessary libraries</h1>

```python
import cv2
from matplotlib import pyplot as plt
```

<h1>Loading image and showing</h1>

```python
image = cv2.imread('pessoas.jpg')
```

```python
# Showing image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()
```

<h1>Creating detector of body</h1>

```python
detector = cv2.CascadeClassifier('fullbody.xml')
```

<h1>Making detections and showing them on the screen.</h1>

```python
detections = detector.detectMultiScale(gray_image, scaleFactor=1.1, minSize=(50,50))
```

```python
for x, y, l, a in detections:
    image = cv2.rectangle(image, (x, y), (x + l, y + a), (0, 255, 0), 2)
plt.imshow(image)
```
